### Проанализировать датафрейм статистики NBA. (файл nba.csv)


Вывести:
1. Проранжировать бюджет команд на зарплату от самого высокого до самого низкого.
2. 10 наиболее высокооплачиваемых игроков в лиге.
3. 3 наиболее высооплачиваемых игроков в каждой команде.
4. Предложить первые пятерки каждой команды на основе их зарплаты (учитывайте, 
   что в первой пятерке на каждой позиции один игрок (см. рисунок)).
5. Самая молодая и самая возростная команда в лиге.
6. Наиболее часто используемый номер игрока в лиге.
7. Имя и рост самого высокого и самого низкого игрока в лиге в метрической системе.
   
P.S. Возможные пропуски в зарплатной ведомости заполнить медианным значением зарплаты команды игрока с пропуском.

```python
import pandas as pd
```

Чтение данных из csv-файла.

```python
nba_data = pd.read_csv('nba.csv')
```

Удаление строк, в которых значения во всех столбцах - NaN.

```python
nba_data.dropna(inplace=True, how='all')
```

Заполнение пустых значений Salary медианными значениями зарплаты в команде.

```python
nba_data['Salary'] = nba_data.groupby('Team')['Salary'].apply(lambda x: x.fillna(x.median()))
nba_data
```

1. Функция groupby  разбивает данные о сумме зарплат, вычисляемые при помомщи sum, по командам, после чего они сортируются по убыванию с помощью sort_values.

```python
print('Бюджет команд на зарплату игрокам (по убыванию):\n', 
      nba_data.groupby('Team')['Salary'].sum().sort_values(ascending=False))
```

2. Функция max используется для вычисления максимальной зарплаты, а функция head(10) оставляет только топ-10 (отсортированных функцией sort_values) значений.

```python
print('Топ самых высокооплачиваемых игроков в лиге:\n', 
      nba_data.groupby('Name')['Salary'].max().sort_values(ascending=False).head(10))
```

Также можно выделить 10 наибольших значений зарплаты в таблице функцией nlargest. 

```python
nba_data.nlargest(10, ['Salary']) 
```

3. Наиболее высооплачиваемых игроков в каждой команде можно найти описанным выше способом, используя цикл для каждой команды.

```python
nba_data.groupby('Team').apply(lambda x: x.nlargest(3, 'Salary'))
```

4. Анологично предыдущему пункту можно выбрать одно наибольшее значение зарплаты для группировки по команде и позиции.

```python
nba_data.groupby(['Team', 'Position']).apply(lambda x: x.nlargest(1, 'Salary'))
```

5. Самая молодая/возрастная команда находится через среднее значение возраста команды и сортировкой по возрастанию/убыванию.

```python
print('Самая молодая команда в лиге:\n', 
      nba_data.groupby('Team')['Age'].mean().sort_values().head(1))
```

```python
print('Самая возрастная команда в лиге:\n', 
      nba_data.groupby('Team')['Age'].mean().sort_values(ascending=False).head(1))
```

6. Наиболее часто используемый номер игрока в лиге - мода всех номеров.

```python
print('Наиболее часто используемый номер игрока в лиге:\n', int(nba_data['Number'].mode()))
```

7. Данные в столбце Height необходимо разделить для перевода в метрическую систему. 
Для этого создан новый столбец Height_cm.

```python
nba_height = nba_data['Height'].str.split('-', expand = True)
nba_data['Height_cm'] = nba_height[0].astype(int) * 30.48 + nba_height[1].astype(int) * 2.54
```

```python
print('Самый высокий игрок в лиге:\n', 
      nba_data.groupby('Name')['Height_cm'].max().sort_values(ascending=False).head(1))
```

```python
print('Самый низкий игрок в лиге:\n', 
      nba_data.groupby('Name')['Height_cm'].max().sort_values().head(1))
```
